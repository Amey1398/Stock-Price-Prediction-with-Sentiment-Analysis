Plotting the RMSE MAPE values for Multivariate LSTM model with different features

```python
import matplotlib.pyplot as plt

# Data from the table
features = ['Only Stocks', 'Stocks + Tweet', 'Stocks + Reddit', 'Stocks + News_Vader', 'Stocks + News_Finbert',
            'Stocks + Tweet + Reddit + News_Vader', 'Stocks + all_compound']
rmse_values = [0.042, 0.031, 0.033, 0.07, 0.08, 0.30, 0.034]
mape_values = [0.031, 0.023, 0.024, 0.13, 0.14, 0.98, 0.025]
```

```python
# Plotting the bar graph
fig, ax = plt.subplots(figsize=(10, 6))

# Bar width can be adjusted as needed
bar_width = 0.35

# Plotting RMSE values
bar1 = ax.bar(features, rmse_values, width=bar_width, label='RMSE', color='b')

# Adding an offset for the next set of bars
bar2_x = [x + bar_width for x in range(len(features))]
bar2 = ax.bar(bar2_x, mape_values, width=bar_width, label='MAPE', color='g')

# Adding labels and title
ax.set_xlabel('Features')
ax.set_ylabel('Values')
ax.set_title('RMSE and MAPE Values for Different Features')
ax.set_xticks([x + bar_width / 2 for x in range(len(features))])
ax.set_xticklabels(features, rotation=45, ha='right')  # Adjust rotation for better readability
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Line chart for RMSE
ax1.plot(features, rmse_values, marker='o', color='b', label='RMSE')
ax1.set_ylabel('RMSE')
ax1.set_xticks(range(len(features)))
ax1.set_xticklabels(features, rotation=45, ha='right')  # Adjust rotation for better readability
ax1.legend()

# Line chart for MAPE
ax2.plot(features, mape_values, marker='s', color='g', label='MAPE')
ax2.set_ylabel('MAPE')
ax2.set_xlabel('Features')
ax2.set_xticks(range(len(features)))
ax2.set_xticklabels(features, rotation=45, ha='right')  # Adjust rotation for better readability
ax2.legend()

# Adding a title
plt.suptitle('RMSE and MAPE Values for Different Features')

# Adjust layout for better spacing
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the table
data = {
    'Features': ['Only Stocks', 'Stocks + Tweet', 'Stocks + Reddit', 'Stocks + News_Vader', 'Stocks + News_Finbert',
                 'Stocks + all_sentiments'],
    'RMSE': [0.042, 0.031, 0.033, 0.071, 0.083, 0.034],
    'MAPE': [0.031, 0.023, 0.024, 0.134, 0.147, 0.025]
}

df = pd.DataFrame(data)

# Sort DataFrame based on high RMSE and MAPE values
df_sorted = df.sort_values(by=['RMSE', 'MAPE'], ascending=[False, False])

# Plotting the bar graph
fig, ax = plt.subplots(figsize=(10, 6))

# Bar width can be adjusted as needed
bar_width = 0.35

# Plotting RMSE values
bar1 = ax.bar(df_sorted['Features'], df_sorted['RMSE'], width=bar_width, label='RMSE', color='b')

# Adding an offset for the next set of bars
bar2_x = [x + bar_width for x in range(len(df_sorted['Features']))]
bar2 = ax.bar(bar2_x, df_sorted['MAPE'], width=bar_width, label='MAPE', color='g')

# Adding labels and title
ax.set_xlabel('Features')
ax.set_ylabel('Values')
ax.set_title('RMSE and MAPE Values for Different Features')
ax.set_xticks([x + bar_width / 2 for x in range(len(df_sorted['Features']))])
ax.set_xticklabels(df_sorted['Features'], rotation=45, ha='right')  # Adjust rotation for better readability
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
